```python
import pandas as pd

url = "https://www.england.nhs.uk/statistics/wp-content/uploads/sites/2/2025/04/Monthly-AE-March-2025.csv"

df = pd.read_csv(url)

print("Shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# Check the Period column values and data types
print("Period unique values (sample):")
print(df['Period'].unique()[:5])

print("\nData types:")
print(df.dtypes)

print("\nNull counts:")
print(df.isnull().sum())

print("\nSample of Org Codes (first 10):")
print(df[['Org Code', 'Org name', 'Parent Org']].head(10).to_string())
```
