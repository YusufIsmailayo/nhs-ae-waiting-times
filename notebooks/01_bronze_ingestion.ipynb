```python
ingestion_log = []
ingestion_timestamp = datetime.now().isoformat()

for file_key, url in files.items():
    output_path = os.path.join(BRONZE_DIR, f"{file_key}.parquet")
    
    try:
        # Download
        df = pd.read_csv(url)
        
        # Add audit columns — Bronze rule: record exactly where this came from
        df['_ingestion_timestamp'] = ingestion_timestamp
        df['_source_url'] = url
        df['_source_file'] = file_key
        
        # Save as Parquet
        df.to_parquet(output_path, index=False)
        
        ingestion_log.append({
            'file': file_key,
            'status': 'SUCCESS',
            'rows': len(df),
            'columns': len(df.columns),
            'url': url
        })
        print(f"✅ {file_key} — {len(df)} rows")
        
    except Exception as e:
        ingestion_log.append({
            'file': file_key,
            'status': 'FAILED',
            'rows': 0,
            'columns': 0,
            'error': str(e),
            'url': url
        })
        print(f"❌ {file_key} — FAILED: {e}")

print(f"\nDone. {sum(1 for l in ingestion_log if l['status']=='SUCCESS')} succeeded, "
      f"{sum(1 for l in ingestion_log if l['status']=='FAILED')} failed.")
```

```python
# Let's find the correct URLs by fetching the archive pages
import urllib.request
import re

def get_csv_urls_from_page(page_url):
    req = urllib.request.Request(page_url, headers={'User-Agent': 'Mozilla/5.0'})
    with urllib.request.urlopen(req) as response:
        html = response.read().decode('utf-8')
    # Find all CSV links
    csv_links = re.findall(r'https://www\.england\.nhs\.uk/statistics/wp-content/uploads/sites/2/[^"]*\.csv', html)
    return csv_links

# Fetch 2022-23 page
print("=== 2022-23 CSV links ===")
links_2022_23 = get_csv_urls_from_page(
    "https://www.england.nhs.uk/statistics/statistical-work-areas/ae-waiting-times-and-activity/ae-attendances-and-emergency-admissions-2022-23/"
)
for l in links_2022_23:
    print(l)
```
